```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi'] = 200
```

```python
ir_exp    = np.loadtxt("azobenzene_ir-spectra_gas-phase.dat")
ir_wB97XD = np.loadtxt("Azobenzene_cis_wB97XD_def2-TZVPP_Freq_G16_23Sep25-0936_ir.txt")
```

```python
plt.plot(ir_exp[:,0],
         ir_exp[:,1],
         "-",
         linewidth=2,
         label="IR Experimental Spectrum in Gas-Phase")
plt.xlabel("Wavelength cm$^{-1}$")
plt.ylabel("Absorbance")
plt.title("Some title of your choice")
plt.legend()
plt.show()


plt.plot(ir_wB97XD[:,0],
         ir_wB97XD[:,1],
         "-",
         linewidth=2,
         label="IR from wB97XD cis - Unscaled")
# This the range of y-axis, might need to be adjusted
#plt.ylim([0.0,0.020])
# You can use LaTeX math syntax in label text by enclosing it in $$
plt.xlabel("Wavelength cm$^{-1}$")
plt.ylabel("Absorbance")
plt.title("Some title of your choice")
plt.legend()
# You can add a plt.savefig(..) call to save the plot to a PNG file.
# adding this command will show the current plot
plt.show()
```

```python
plt.plot(ir_exp[:,0],
         ir_exp[:,1],
         "-",
         linewidth=2,
         label="IR Experimental Spectrum in Gas-Phase")
plt.xlabel("Wavelength cm$^{-1}$")
plt.ylabel("Absorbance")
plt.title("Some title of your choice")
plt.legend()

AbsorbanceScaling_Gaussian=1./2700.

plt.plot(ir_wB97XD[:,0],
         AbsorbanceScaling_Gaussian*ir_wB97XD[:,1],
         "-",
         linewidth=2,
         label="IR from wB97XD cis - Unscaled")
# This the range of y-axis, might need to be adjusted

# You can use LaTeX math syntax in label text by enclosing it in $$
plt.xlabel("Wavelength cm$^{-1}$")
plt.ylabel("Absorbance")
plt.title("Some title of your choice")
plt.legend()
# You can add a plt.savefig(..) call to save the plot to a PNG file.
# adding this command will show the current plot

plt.ylim([0.0,0.04])
plt.xlim([0.0,4000])

plt.savefig("plot.png")

plt.show()
```
